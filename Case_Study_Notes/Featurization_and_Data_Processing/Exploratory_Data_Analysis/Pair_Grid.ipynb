```python
import pandas as pd
import seaborn as sns
```

Following https://seaborn.pydata.org/tutorial/axis_grids.html

```python
crab_data = pd.read_csv("CrabAgePrediction.csv")
```

```python
g = sns.PairGrid(crab_data)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

This gives us plots for the continuous valued columns

We can get information about the categorical columns "by hand"

```python
crab_data.value_counts()
```

We see that the only truly categorical column is "Sex".

```python
sns.histplot(crab_data["Sex"])
```

```python
sns.scatterplot(crab_data["Sex"], crab_data["Weight"])
```

```python
sns.scatterplot(x=crab_data["Sex"], y=crab_data["Height"])
```
